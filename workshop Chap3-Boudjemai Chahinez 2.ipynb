```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv(r"C:/Users/Chahinez/Downloads/train.csv")
df.head()

```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'].dropna(), bins=30, color='skyblue')
plt.title("Distribution des âges")
plt.xlabel("Âge")
plt.ylabel("Nombre de passagers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Pclass'].value_counts().plot(kind='bar', color='orange')
plt.title("Répartition des classes")
plt.xlabel("Classe")
plt.ylabel("Nombre de passagers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Sex', y='Age', data=df)
plt.title("Âge selon le sexe")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Survived']], hue='Survived')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap des corrélations")
plt.show()
```

```python
fig = px.scatter(df, x='Age', y='Fare', color='Survived', hover_data=['Name'])
fig.show()
```

```python
fig = px.histogram(df, x='Pclass', color='Survived', barmode='group')
fig.show()
```
